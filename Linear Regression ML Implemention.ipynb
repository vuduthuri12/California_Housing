```python
#dividing independent and dependent features
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python

Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 42)
```

```python
X_test
```

```python
X_train
```

```python
#standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler, open('scaler.pkl', 'wb'))
print(" Scaler saved as scaler.pkl")
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,Y_train)
```

```python

print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
#on which parameters model has been trained
reg.get_params()
```

```python
reg_pred = reg.predict(X_test)
```

```python
reg_pred
```

```python
plt.scatter(Y_test,reg_pred)
```

```python
#residuals means error
res = Y_test - reg_pred
```

```python
res
```

```python
#plot resiuals
sns.displot(res,kind='kde')
```

```python
#scatter plot with prediction and residuals
plt.scatter(reg_pred,res)
```

```python

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(Y_test,reg_pred))
print(mean_squared_error(Y_test,reg_pred))
print(np.sqrt(mean_squared_error(Y_test,reg_pred)))
```

R Squre and Adjusted R Square

Formula 
R^2 = 1-SSR/SST
R^2 = coefficient of determination SSR = Sum of Squares of residuas sst = total sum of squares

```python
from sklearn.metrics import r2_score
score = r2_score(Y_test,reg_pred)
print(score)
```

ADjusted R^2 = 1-[(1-R^2)*(n-1)/(n-k-1)]
where
R2 = The R2 of the model n:the no of obsevations k:the no of predictor variable

```python
1-(1-score)*(len(Y_test-1))/(len(Y_test)-X_test.shape[1]-1)
```
